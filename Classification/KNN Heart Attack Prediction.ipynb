```python
# A K-Nearest Neighbors classifier applied to a dataset containing medical infomation about hospital patients.
# Each feature represents to a medical test, while each instance corresponds to a patient and their respective score
# on each test. The target attribute was whether the patient had suffered from a heart attack or not. The goal was to
# predict patients who had suffered from a heart attack based on their test results. This technique could then be used
# on future patients to idenfity those who were at risk of having a heart attack. Data obtained from 'www.kaggle.com'. 

# 1. Load the dataset as Pandas DataFrame object. Create another object containing just the target attribute.

import pandas
heart_data=pandas.read_csv('heart.csv')
heart_data2=heart_data.copy()
heart_data2=heart_data2.drop(['age','sex','cp','trestbps','chol','fbs',
                              'restecg','thalach','exang','oldpeak','slope','ca','thal'], axis=1)
```

```python
# 2. Prepare the data by performing normalization. The target attribute is numeric (1 for 'suffered heart attck', 0 if not),
# and so is included in the normalization. To correct this, the object created in step 1 which contains just the target
# attribute is used to replace the normalized target attribute. The correlation of all the attributes is then calculated.
# The target attribute is shown in column 13. Features that have a low correleation with the target attribute are removed
# from the dataset. 

import sklearn.preprocessing
heart_data3=heart_data.copy()
norm_data=sklearn.preprocessing.normalize(heart_data3)
norm_data=pandas.DataFrame(norm_data)
norm_data=norm_data.drop(13, axis=1)
norm_data.insert(13,'13', heart_data2)
norm_data.corr()
```

```python
norm_data=norm_data.drop(labels=[0,1,3,4,5,6,7,10,11,12], axis=1)
```

```python
# Split the dataset into a training set and a test set. 

import sklearn.model_selection
train_set, test_set = sklearn.model_selection.train_test_split(norm_data, test_size=0.2)
train_set_att=train_set.copy()
train_set_targ=train_set.copy()
train_set_att=train_set_att.drop(labels='13', axis=1)
train_set_targ=train_set_targ.drop(labels=[2,8,9], axis=1)
test_set_att=test_set.copy()
test_set_targ=test_set.copy()
test_set_att=test_set_att.drop(labels='13', axis=1)
test_set_targ=test_set_targ.drop(labels=[2,8,9], axis=1)
```

```python
# Apply the K-Nearest Neighbors algorithm and use it to make predictions on the test set.

import sklearn.neighbors
KNN=sklearn.neighbors.KNeighborsClassifier(n_neighbors=9)
KNN.fit(train_set_att, train_set_targ)
predictions=KNN.predict(test_set_att)
```

```python
# Compare the precticed values with the true values.  

import sklearn.metrics
accuracy=round(sklearn.metrics.accuracy_score(test_set_targ, predictions),2)
print('Model accuracy =',accuracy)
```
